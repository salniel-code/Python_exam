```python
# Lösning uppgift 1a
# ------------------

# input och listor
N = int(input("Ange antalet N: "))
list = []
temp_list = []

# nästlade loopar a, b från 1 till N och c från N till 1.
for a in range(1, N):
    for b in range(1, N):
        for c in range(N, 1, -1):
            # om a^2 + b^2 = c^2, tilldela temp_list med a,b,c. lägg in temp_list i list
            if a**2+b**2 == c**2:
                temp_list = [a,b,c]
                list.append(temp_list)
i = 0
temp = []
# iterera 0 till längden av lista. tilldela temp med listan från indexplats "i" från list
while i < len(list):
    temp = list[i]
    # iterera genom list. om  kombinationen (temp) redan finns, ta bort från list
    for row in list:
        if temp[0] == row[1] and temp[1] == row[0]:
            list.remove(row)
    i = i + 1
# skriv ut antal kombinationer genom att ta längd av list. iterera och skriv ut varje rad av list.
amount = len(list)
print(f'Det finns {amount} heltalskombinationer som uppfyller Pythagoras sats:')
count = 0
for row in list:
    count = count + 1
    print(count, ". ", row)
```

```python
# Lösning uppgift 1b
# ------------------
def omvandla(tal):
    bin_str = ''
    dict = {'0': '0000', '1': '0001', '2': '0010', '3': '0011', '4': '0100', '5':
'0101', '6': '0110', '7': '0111', '8': '1000', '9': '1001', 'A': '1010',
'B': '1011', 'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'}
    #för varje karaktär i "tal", om karaktären finns i lexikon, lägg till i sträng bin_str och returnera denna
    for char in tal:
        if char in dict:
            bin_str = bin_str + dict[char]
    return bin_str    
# ber användare om input, skriver ut med att anropa funktion omvandla(tal) med användarens input som argument
user_input = input('Ange ett hexadecimalt tal som ska omvandlas: ')
print(f'Det binära talet blir: {omvandla(user_input)}')
```

```python
# Lösning uppgift 2b
# ------------------
def sort_funktion(in_lista):                                            
    n=len(in_lista)
    for i in range(n):
        for j in range(n-i-1):
            if in_lista[j] > in_lista[j+1]:
                in_lista[j], in_lista[j+1] = in_lista[j+1], in_lista[j]
    return in_lista

test_lista= [6, 37, 25, 12, 4, 1, 9]
sort_funktion(test_lista)
# Ändringar:
# rad 4: j in range(0,n) till range(n-i-1)
# rad 5: if in_lista[j-1] till in_lista[j]
# rad 6: in_lista[j-1] till in_lista[j+1]
# rad 6: == till = 
# rad 7: return i funktion från in_list till in_lista
```

```python
# Lösning uppgift 3a
# ------------------
import csv
# Funktion som läser csv fil, lägger varje rad i data och returnerar data (innehållet) som lista
def read_file(file_name):
  with open(file_name, mode = 'r', encoding = 'UTF-8') as file:
    csv_reader = csv.reader(file, delimiter=';')
    data = []
    for line in csv_reader:
      data.append(line)
    return data
# Tilldelar kpiData innehållet från filen kpi.csv, 
kpiData = read_file('kpi.csv')
# Loopar 3 ggr för utskrift av 3 rader i kpiData, indexplats = [i]
i = 0
while i < 3:
    print(kpiData[i])
    i = i + 1
```

```python
# Lösning uppgift 3b
# ------------------

# Tilldelar kpiData innehållet från filen kpi.csv, 
kpiData = read_file('kpi.csv')

# Funktion beräknar summa av talen, räknar antal tal, returnerar medelvärde = sum/count            
def medel(lista):
    n = len(lista)
    sum = 0
    count = 0
    for i in range(n):
        sum = sum + float(lista[i])
        count = count + 1
    return sum / count
# Funktion, min_v tilldelas första värdet i lista, ersätts om nästa värde är mindre. returnerar minsta värde
def min_value(lista):
    n = len(lista)
    min_v = lista[0]
    for i in range(n):
        if min_v > lista[i]:
            min_v = lista[i]
    return min_v
# Funktion, min_v tilldelas första värdet i lista, ersätts om nästa värde är mindre. returnerar minsta värde
def max_value(lista):
    n = len(lista)
    max_v = lista[0]
    for i in range(n):
        if max_v < lista[i]:
            max_v = lista[i]
    return max_v

data = []
years = []

# inkrementerar genom listan för att lägga in årens värde från tidigast till senast i data (sortera). tar bort årtal från rad. lägg årtal i en lista.
for n in range(1980, 2024):
    for row in kpiData:
        if str(n) == row[0]:
            row.remove(str(n))
            # tar bort tomma strängar ur rader. lägger in år (n) i years, row i data
            while ' ' in row:
                row.remove(' ')
            while '' in row:
                row.remove('')
            years.append(n)
            data.append(row)

inflat_yrs = []
i = 0
# itererar # längden av data, beräknar inflationstakt genom medelvärde av element [i] och element [i-1]
while i < len(data)-1:
    # OBS - använder den angivna formeln, dock kommer detta bli snett tillsammans med de årtal där mina årtal blir -1
    # i jämförelse med de i uppgiftsbesk. tabell. 
    x = medel(data[i+1])
    y = medel(data[i])
    z = 100
    inflat = round(((x -y)/y)*z,1)
    inflat_yrs.append(inflat)
    i = i + 1

# anropar min_value(), max_value(), med inflat_yrs som argument och tilldelar variabel returnerat värde
min_inflat = min_value(inflat_yrs)
max_inflat = max_value(inflat_yrs)

table_list = []
# itererar för att hitta index nummer och tillhörande årtal
i = 0
while i < len(inflat_yrs)-1:
    # jämför max_inflat med värde i inflat_yrs, om sant- lägg värdet och tillhörande årtal i lista
    if max_inflat == inflat_yrs[i]:
        table_list.append(max_inflat)
        table_list.append(years[i])
    i = i +1
i = 0
while i < len(inflat_yrs)-1:
    # jämför min_inflat med värde i inflat_yrs, om sant- lägg värdet och tillhörande årtal i lista
    if min_inflat == inflat_yrs[i]:
        table_list.append(min_inflat)
        table_list.append(years[i])
    i = i +1
# Gör till funktion för kunna anropa i meny uppgift
def tabell_ar():
    # utskrift av tabell, inkl table_list element
    print('-'*100)
    print('Inflationen i Sverige under tidsperioden 1 januari 1980 -- september 2023 uttryckt i %')
    print('-'*100) 
    print(f'                      Årlig                                Månad                         Medelinflation')
    print(f'                      Högst   år    Lägst   år    Högst  år /mån    Lägst   år /mån      under perioden %')
    print('-'*106)
    print(f'inflationstakt årlig: {table_list[0]}   {table_list[1]}   {table_list[2]}   {table_list[3]}')
tabell_ar()
# Tilldelar max och min inflationstakt tillhörande år till variabler för nästa uppgift
max_year = table_list[1]
min_year = table_list[3]

temp = []
list = []
i = 0
# Lägger rad i temporär lista, iterar den och lägger in dess element i list för att ha alla element/värden i en lista
while i < len(data):
    temp = data[i]
    for elements in temp:
        list.append(float(elements))
    i = i + 1   

inflat_months = []
i = 0
# itererar # längden av list, beräknar inflationstakt genom medelvärde av element [i] och element [i-1]
while i < len(list)-1:
    x = list[i+1]
    y = list[i]
    z = 100
    inflat = round(((x -y)/y)*z,2)
    inflat_months.append(inflat)
    i = i + 1

# tilldelar returnerat värde av anropad funktion, som tar argument inflat_months
min_inflat = min_value(inflat_months)
max_inflat = max_value(inflat_months)

years_list = []
# lägger in 12 (månader) av varje instans av yr (år), för att sedan lokalisera rätt årtal för min, max _inflat
for yr in years:
    i = 0
    while i <= 12:
        years_list.append(yr)
        i = i + 1

# Lägger till månader # yrs i months för att lokalisera rätt månad till inflationstakt
yrs = 2024 - 1980
months_list = yrs*(kpiData[0][1:13])
        
table_list = []

i = 0 
# iterar längden av inflat_months. jämför max_inflat med värde i inflat_months, om sant- lägg värdet,tillhörande årtal, månad i lista
while i < len(inflat_months):
    if max_inflat == inflat_months[i]:
        table_list.append(max_inflat)
        table_list.append(years_list[i])
        table_list.append(months_list[i])

    i = i + 1
# iterar längden av inflat_months. jämför min_inflat med värde i inflat_months, om sant- lägg värdet,tillhörande årtal, månad i lista
i = 0
while i < len(inflat_months):
    if min_inflat == inflat_months[i]:
        table_list.append(min_inflat)
        table_list.append(years_list[i])  
        table_list.append(months_list[i])
    i = i + 1
# Beräknar medelinflationstakt för hela perioden. lägger till i table_list
medel_all_inflat = round(medel(inflat_months),1)
table_list.append(medel_all_inflat)
# Gör till funktion för kunna anropa i meny uppgift
def tabell_man():
    # utskrift av tabell, inkl table_list element
    print(f'inflationstakt månatlig:                          {table_list[0]}   {table_list[1]}/{table_list[2]}   {table_list[3]}   {table_list[4]}/{table_list[5]}        {table_list[6]}')
tabell_man()
```
